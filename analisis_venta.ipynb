```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Producto": ["A", "B", "C", "D", "E"],
    "Ventas": [1200, 850, 430, 2100, 1600]
}

df = pd.DataFrame(data)
df

ventas_np = np.array(df["Ventas"])

promedio = np.mean(ventas_np)
maximo = np.max(ventas_np)
minimo = np.min(ventas_np)

print("Ventas:", ventas_np)
print("Promedio de ventas:", promedio)
print("Venta máxima:", maximo)
print("Venta mínima:", minimo)

sns.barplot(x=df["Producto"], y=df["Ventas"])
plt.title("Ventas por Producto")
plt.show()
```

### Explicación

- **np.mean()**: Calcula el promedio de las ventas.
- **np.max()**: Obtiene el valor más alto.
- **np.min()**: Obtiene el valor más bajo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data ={
    'Producto': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Ventas': [120, 340, 290, 410, 150, 380],
    'Costo': [80, 200, 180, 260, 90, 240],
    'Categoria': ['Tecnología', 'Hogar', 'Tecnología', 'Hogar', 'Oficina', 'Tecnología'],
    'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo']
}

df = pd.DataFrame(data)
df['Ganancia'] = df['Ventas'] - df['Costo']
df.head()

productos_ganancia_mayor_100 = df[df['Ganancia'] > 100]
productos_ganancia_mayor_100

plt.figure()
plt.bar(df['Producto'], df['Ganancia'])
plt.title('Ganancia por Producto')
plt.xlabel('Producto')
plt.ylabel('Ganancia')
plt.legend(['Ganancia'])

filtro = (df['Categoria'] == 'Tecnología') & (df['Ventas'] > 300)
df_filtrado = df.loc[filtro, ['Producto', 'Ventas', 'Ganancia']]
df_filtrado
```

```python
resumen_mes = df.groupby('Mes').agg({
    'Ventas': 'sum',
    'Ganancia': 'mean'
}).reset_index()
resumen_mes
```

## Diferencia entre `sum()` y `mean()`

- **`sum()`**: Suma todos los valores de una columna o grupo.  
- **`mean()`**: Calcula el valor promedio (suma de los valores dividido entre la cantidad de elementos).

```python
plt.figure(figsize=(8,5))
grafico = sns.barplot(x='Producto', y='Ventas', data=df, palette='Blues_d', dodge=False)
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(df['Producto'], df['Ventas'], marker='o', linestyle='-', color='blue', label='Ventas')
plt.plot(df['Producto'], df['Costo'], marker='s', linestyle='--', color='red', label='Costo')

plt.title('Comparación Ventas vs Costo por Producto', fontsize=14)
plt.xlabel('Producto', fontsize=12)
plt.ylabel('Monto', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

## Explicación de la tendencia

En el gráfico se observa que las **ventas superan al costo en todos los productos**, lo que indica que todos generan ganancia.  

Además, los productos **D** y **F** presentan la mayor diferencia entre ventas y costo, mostrando que son los **más rentables**.  

Por otro lado, los productos **A** y **E** tienen menores ventas y ganancia, por lo que son los **menos destacados económicamente**.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(x='Categoria', y='Ventas', data=df, palette='Set2', dodge=False)
plt.title('Distribución de Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.show()
```

## Análisis de la variabilidad en ventas por categoría

Al observar el boxplot, se nota que la categoría **Tecnología** presenta la **mayor variabilidad** en ventas.  

Esto se evidencia porque el rango intercuartílico y los valores extremos (outliers) de Tecnología son más amplios en comparación con las categorías **Hogar** y **Oficina**, lo que indica que las ventas dentro de esta categoría varían más significativamente entre los productos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Ventas', y='Ganancia', hue='Categoria', s=100)

plt.title('Relación entre Ventas y Ganancia por Categoría', fontsize=14)
plt.xlabel('Ventas', fontsize=12)
plt.ylabel('Ganancia', fontsize=12)
plt.legend(title='Categoría')
plt.show()
```

## Relación entre Ventas y Ganancia

En el scatterplot entre **Ventas** y **Ganancia**, se observa una relación positiva entre ambas variables: a medida que las ventas aumentan, también lo hace la ganancia. Esto tiene sentido, ya que la **ganancia** se calcula como la diferencia entre **ventas** y **costo**. Por lo tanto, productos con mayores ventas generalmente generan mayores ganancias.

### Observaciones por Categoría:
- Los productos de la categoría **Tecnología** parecen tener una mayor dispersión en cuanto a **ventas** y **ganancia**, indicando que algunos productos dentro de esta categoría tienen ventas altas pero también costos elevados, lo que afecta la ganancia.
- Los productos en las categorías **Hogar** y **Oficina** muestran una relación más constante, con menos variabilidad en las ganancias a pesar de la variación en las ventas.

### Conclusión:
Los productos con **ventas más altas** tienen, en general, **ganancias más altas**, aunque también existen algunos productos con ventas altas pero ganancias moderadas debido a costos elevados. Esto muestra que las ganancias dependen no solo de las ventas, sino también de los costos asociados a cada producto.
